```python
import pandas as pd

# Dataset 1: Student Personal Information
data1 = {
    'Student_ID': [101,102,103,104],
    'Name': ['Asha','Ravi','Neha','Aman'],
    'Age': [20,21,22,23],
    'Gender': ['F','M','F','M']
}

df1 = pd.DataFrame(data1)

# Dataset 2: Student Academic Records
data2 = {
    'Student_ID': [101,102,103,104],
    'Course': ['AI','Data Science','Machine Learning','AI'],
    'Marks': [85,78,90,88]
}

df2 = pd.DataFrame(data2)

# Integrate datasets
df_final = pd.merge(df1, df2, on='Student_ID')

print("Integrated Dataset:")
print(df_final)
```

```python
average_marks = df_final['Marks'].mean()

print("Average Marks of Students:", average_marks)
```
