```python
from scipy import stats
import numpy as np

data = [12, 15, 14, 10, 13, 18, 16]

# Mean using scipy
mean = stats.tmean(data)
print("Mean using scipy:", mean)

# One-sample t-test (compare mean with 12)
t_stat, p_value = stats.ttest_1samp(data, 12)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```
